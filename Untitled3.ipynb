```python
x = data.drop('target', axis=1)
y = data['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train
y_train
x_test
y_test
```

```python
#1. Lasso Feature Selection
lasso = Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
lasso_features = x.columns[lasso.coef_ != 0].tolist()
lasso_model = LogisticRegression()
lasso_model.fit(x_train[lasso_features], y_train)
lasso_pred = lasso_model.predict(x_test[lasso_features])
lasso_accuracy = accuracy_score(y_test, lasso_pred)
print(f"Lasso Feature Selection Accuracy: {lasso_accuracy:.4f}")

# 2. Ridge Feature Selection
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
ridge_features = x.columns[ridge.coef_ != 0].tolist()
ridge_model = LogisticRegression()
ridge_model.fit(x_train[ridge_features], y_train)
ridge_pred = ridge_model.predict(x_test[ridge_features])
ridge_accuracy = accuracy_score(y_test, ridge_pred)
print(f"Ridge Feature Selection Accuracy: {ridge_accuracy:.4f}")

# 3. Correlation-Based Feature Selection
correlation_matrix = data.corr()
high_corr_features = correlation_matrix['target'][correlation_matrix['target'].abs() > 0.5].index.tolist()
high_corr_features.remove('target')
corr_model = LogisticRegression()
corr_model.fit(x_train[high_corr_features], y_train)
corr_pred = corr_model.predict(x_test[high_corr_features])
corr_accuracy = accuracy_score(y_test, corr_pred)
print(f"Correlation-Based Feature Selection Accuracy: {corr_accuracy:.4f}")

# 4. Recursive Feature Elimination (RFE) Selection
rfe = RFE(estimator=LogisticRegression(), n_features_to_select=5)
rfe.fit(x_train, y_train)
rfe_features = x.columns[rfe.support_].tolist()
rfe_model = LogisticRegression()
rfe_model.fit(x_train[rfe_features], y_train)
rfe_pred = rfe_model.predict(x_test[rfe_features])
rfe_accuracy = accuracy_score(y_test, rfe_pred)
print(f"RFE Feature Selection Accuracy: {rfe_accuracy:.4f}")

# 5. Mutual Information Feature Selection
mi_scores = mutual_info_classif(x_train, y_train)
mi_series = pd.Series(mi_scores, index=x.columns).sort_values(ascending=False)
top_mi_features = mi_series.index[:5].tolist()
mi_model = LogisticRegression()
mi_model.fit(x_train[top_mi_features], y_train)
mi_pred = mi_model.predict(x_test[top_mi_features])
mi_accuracy = accuracy_score(y_test, mi_pred)
print(f"Mutual Information Feature Selection Accuracy: {mi_accuracy:.4f}")

#6. Alternative mRMR Approximation
top_features = mi_series.index[:10].tolist()
selected_features = []
correlation_threshold = 0.7

for feature in top_features:
    if all(x_train[feature].corr(x_train[selected]) < correlation_threshold for selected in selected_features):
        selected_features.append(feature)

# Train model on selected features and evaluate
mrmr_approx_model = LogisticRegression()
mrmr_approx_model.fit(x_train[selected_features], y_train)
mrmr_approx_pred = mrmr_approx_model.predict(x_test[selected_features])
mrmr_approx_accuracy = accuracy_score(y_test, mrmr_approx_pred)
print(f"Approximate mRMR Feature Selection Accuracy: {mrmr_approx_accuracy:.4f}")
```

```python
# Define the ANN model
def create_ann(input_shape):
    model = Sequential()
    model.add(Dense(64, input_dim=input_shape, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Define the CNN model
def create_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(64, kernel_size=2, activation='relu', input_shape=(input_shape, 1)))
    model.add(Flatten())
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Define the RNN model
def create_rnn(input_shape):
    model = Sequential()
    model.add(SimpleRNN(64, input_shape=(input_shape, 1), activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model


# Initialize dictionary for traditional machine learning models
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Dictionary to store all models
models = {
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'ANN': create_ann(x_train.shape[1]),
    'CNN': create_cnn(x_train.shape[1]),
    'RNN': create_rnn(x_train.shape[1])
}
# Training and accuracy evaluation for each model
for model_name, model in models.items():
    if model_name in ['ANN', 'CNN', 'RNN']:
        # Reshape data for ANN, CNN, and RNN
        if model_name == 'CNN' or model_name == 'RNN':
            x_train_reshaped = np.expand_dims(x_train.values, axis=-1).astype(np.float32)
            x_test_reshaped = np.expand_dims(x_test.values, axis=-1).astype(np.float32)
        else:
            x_train_reshaped = x_train.astype(np.float32)
            x_test_reshaped = x_test.astype(np.float32)

        # Train ANN, CNN, or RNN model
        model.fit(x_train_reshaped, y_train, epochs=10, batch_size=16, verbose=0)

        # Evaluate accuracy on test set
        _, accuracy = model.evaluate(x_test_reshaped, y_test, verbose=0)
        print(f"{model_name} Accuracy: {accuracy:.4f}")

    else:
        # Train traditional ML model
        model.fit(x_train, y_train)

        # Predict and calculate accuracy
        y_pred = model.predict(x_test)
        accuracy = accuracy_score(y_test, y_pred)
        print(f"{model_name} Accuracy: {accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.base import clone
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical

best_accuracy = 0
best_model_name = ""
best_fs_name = ""
best_model = None

# List to store results for comparison
results = []

# Check if models and feature selection methods are not empty
if not models or not feature_selection_methods:
    raise ValueError("Models or Feature Selection Methods are not defined or empty.")

# Ensure that the datasets (x_train, y_train, x_test, y_test) are correctly defined
if x_train is None or x_test is None or y_train is None or y_test is None:
    raise ValueError("Training or test datasets (x_train, x_test, y_train, y_test) are not defined or empty.")

# Loop over models and feature selection methods
for model_name, model in models.items():
    for fs_name, selected_features in feature_selection_methods.items():
        print(f"\nEvaluating {model_name} with {fs_name} feature selection:")

        try:
            # Select the features for training and testing
            x_train_selected = x_train[selected_features].astype(np.float32)
            x_test_selected = x_test[selected_features].astype(np.float32)

            # Initialize variables for tracking model accuracy
            accuracy = 0

            # If it's a neural network model (ANN, CNN, RNN)
            if model_name in ['ANN', 'CNN', 'RNN']:
                # Reshape data for neural networks if necessary
                if model_name == 'ANN':
                    x_train_reshaped = x_train_selected
                    x_test_reshaped = x_test_selected
                elif model_name == 'CNN':
                    x_train_reshaped = np.expand_dims(x_train_selected, axis=-1)
                    x_test_reshaped = np.expand_dims(x_test_selected, axis=-1)
                elif model_name == 'RNN':
                    x_train_reshaped = np.expand_dims(x_train_selected, axis=2)
                    x_test_reshaped = np.expand_dims(x_test_selected, axis=2)

                # Check if output needs to be one-hot encoded (for classification tasks)
                if len(np.unique(y_train)) > 2:
                    y_train_reshaped = to_categorical(y_train)
                    y_test_reshaped = to_categorical(y_test)
                else:
                    y_train_reshaped = y_train
                    y_test_reshaped = y_test

                # Train and evaluate neural networks
                model.fit(x_train_reshaped, y_train_reshaped, epochs=10, batch_size=16, verbose=0)
                _, accuracy = model.evaluate(x_test_reshaped, y_test_reshaped, verbose=0)

            else:
                # For traditional ML models, standardize the data
                scaler = StandardScaler()
                x_train_scaled = scaler.fit_transform(x_train_selected)
                x_test_scaled = scaler.transform(x_test_selected)

                # Cross-validation for better accuracy estimation
                cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
                accuracy = cross_val_score(model, x_train_scaled, y_train, cv=cv, scoring='accuracy').mean()

            print(f"Accuracy: {accuracy:.4f}")

            # Store the results for comparison
            results.append({
                "model_name": model_name,
                "fs_name": fs_name,
                "accuracy": accuracy
            })

            # Update the best model if the current model has a better accuracy
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_model_name = model_name
                best_fs_name = fs_name
                # Make sure to reinitialize the best model to avoid any issues with model modification
                best_model = clone(model)
        except KeyError as e:
            print(f"Error: {e}")
            print(f"Skipping {fs_name} due to missing or incorrect features.")

        except Exception as e:
            print(f"Unexpected error: {e}")
            print(f"Skipping {model_name} with {fs_name} due to a processing error.")

# Ensure best_model is not None
if best_model is None:
    raise ValueError("No best model was selected during evaluation.")

# Print the best model and feature selection method
print("\nBest Model and Feature Selection:")
print(f"Model: {best_model_name}")
print(f"Feature Selection: {best_fs_name}")
print(f"Best Accuracy: {best_accuracy:.4f}")

# Optionally, print all results for comparison
print("\nAll Evaluation Results:")
for result in results:
    print(f"Model: {result['model_name']}, Feature Selection: {result['fs_name']}, Accuracy: {result['accuracy']:.4f}")

# Now, train the best model with the selected features
print(f"\nTraining the best model: {best_model_name} with {best_fs_name} feature selection")

# Select the features based on the best feature selection method
x_train_selected = x_train[feature_selection_methods[best_fs_name]].astype(np.float32)
x_test_selected = x_test[feature_selection_methods[best_fs_name]].astype(np.float32)

# Ensure correct reshaping for neural networks (like CNN)
if best_model_name in ['ANN', 'CNN', 'RNN']:
    # Reshape data for neural networks if necessary
    if best_model_name == 'ANN':
        x_train_reshaped = x_train_selected
        x_test_reshaped = x_test_selected
    elif best_model_name == 'CNN':
        x_train_reshaped = np.expand_dims(x_train_selected, axis=-1)
        x_test_reshaped = np.expand_dims(x_test_selected, axis=-1)  # Add channel dimension
    elif best_model_name == 'RNN':
        x_train_reshaped = np.expand_dims(x_train_selected, axis=2)
        x_test_reshaped = np.expand_dims(x_test_selected, axis=2)

    # Check if output needs to be one-hot encoded (for classification tasks)
    if len(np.unique(y_train)) > 2:  # If more than 2 classes, we need one-hot encoding
        y_train_reshaped = to_categorical(y_train)
        y_test_reshaped = to_categorical(y_test)
    else:
        y_train_reshaped = y_train
        y_test_reshaped = y_test

    # Train and evaluate neural networks
    best_model.fit(x_train_reshaped, y_train_reshaped, epochs=10, batch_size=16, verbose=1)
    _, best_accuracy_final = best_model.evaluate(x_test_reshaped, y_test_reshaped, verbose=1)

else:
    # For traditional ML models, standardize the data
    scaler = StandardScaler()
    x_train_scaled = scaler.fit_transform(x_train_selected)
    x_test_scaled = scaler.transform(x_test_selected)

    # Train the best model
    best_model.fit(x_train_scaled, y_train)

    # Evaluate the model
    best_accuracy_final = best_model.score(x_test_scaled, y_test)

# Output the final accuracy of the best model
print(f"\nFinal Accuracy of {best_model_name} with {best_fs_name}: {best_accuracy_final:.4f}")

# Create a DataFrame for plotting
df = pd.DataFrame(results)

# Set up the plot
plt.figure(figsize=(12, 8))
sns.barplot(x='accuracy', y='model_name', data=df, palette='viridis')

# Mark the best model with a red color (highlighting the best accuracy)
plt.scatter(best_accuracy, df[df['model_name'] == best_model_name].index[0],
            color='red', s=100, label=f"Best Model: {best_model_name} ({best_accuracy_final:.4f})")

# Title and labels
plt.title('Model-Feature Selection Accuracy Comparison', fontsize=16)
plt.xlabel('Accuracy', fontsize=12)
plt.ylabel('Model - Feature Selection Combination', fontsize=12)

# Display the best accuracy value and legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
